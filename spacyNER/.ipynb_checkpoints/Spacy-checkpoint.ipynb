```python
import json 

with open("de_train.json",'r')as f:
    train = json.load(f)
    
with open("de_dev.json",'r')as f:
    dev = json.load(f)
    
train[-1][1].pop(-1)
dev[-1][1].pop(-1)
```

```python
import spacy
from spacy import displacy
nlp1_de = spacy.load('de_core_news_sm')
nlp2_de = spacy.load('de_core_news_md')
```

```python
### Obtenir entités nommées annotées
import string
ponctuation = set([x for x in string.punctuation])
entity_exist = []
for x in train:
    phrase = []
    text = x[0]
    l = x[1]
    for i in l:
        a = i[0]
        b = i[1]
        while text[b]!=' ' and text[b] not in ponctuation:
            b+=1
            if b==len(text):
                break
        t = text[i[0]:b].replace('¬',"")
        phrase.append([t,i[2].upper()])
    entity_exist.append(phrase)
print(len(entity_exist),entity_exist)
```

```python
# Nettoyer les données
#
# 1. supprimer les ponctuations

import string
ponctuation = [x for x in string.punctuation]
for i in range(len(train)):
    text = train[i][0]
    for y in ponctuation:
        text = text.replace(y," ")
    text = text.replace('¬','')
    train[i][0]=text
print(train[1][0])
```

```python
from spacy import displacy
text = train[1][0]
text = text.replace("¬","")
doc1 = nlp1_de(text)
displacy.serve(doc1,style="ent")
```

```python
def show_entity_model1(exemple,i):
    text = exemple[0]
    result = exemple[1]
    entite = nlp1_de(text)
    res = [[ent.text.strip(),ent.start_char,ent.end_char,ent.label_] for ent in entite.ents]
    statistique = stats(res,entity_exist[i])
    return statistique

def stats(res,entity):
    trouvé_correct,incorrect,non= [],[],[]
    entity_list = entity[:]
    entity_set = set([x[0] for x in entity])
    candidat_set = set()
    #print(entity_list," ",entity_set," ",candidat_set)
    for x in res:
        l = x[0].split(" ")
        for y in l:
            candidat_set.add(y)
    #print(candidat_set)
    for x in res:
        l = x[0].split(" ")
        flag = False
        for y in l:
            if y in entity_set:
                trouvé_correct.append([x[0],x[3]])
                flag = True
                break
        if not flag:
            incorrect.append([x[0],x[3]])
    for i in range(len(entity_list)):
        if entity_list[i][0] not in candidat_set:
            non.append(entity_list[i])
    return trouvé_correct,incorrect,non
    

trouvée = []
incorrect = []
non = []
for i in range(len(train)):
    res = show_entity_model1(train[i],i)
    trouvée+=res[0]
    incorrect+=res[1]
    non+=res[2]
    
print(len(trouvée),len(incorrect),len(non))
```

```python
def show_entity_model2(exemple,i):
    text = exemple[0]
    result = exemple[1]
    entite = nlp2_de(text)
    res = [[ent.text.strip(),ent.start_char,ent.end_char,ent.label_] for ent in entite.ents]
    statistique = stats(res,entity_exist[i])
    return statistique

t = []
inc = []
n = []
for i in range(len(train)):
    res = show_entity_model2(train[i],i)
    t+=res[0]
    inc+=res[1]
    n+=res[2]
    
print(len(t),len(inc),len(n))
```

```python
class Solution:
    #. K i. T e
    #.     K     k
    #.   i. I
    #.  t T
    
    #. 
    def __init__(self):
        self.stats = {}
        self.v = {}
        
    def spellchecker(self, wordlist, queries):
        wordset = set(wordlist)
        #print(wordlist.index("zeo")," ",wordlist.index("zuo"))
        voyel = {'a','e','i','o','u','A','E','I','O','U'}
        visited = set()
        for i in range(len(wordlist)):
            tmp = self.trace(wordlist[i],voyel)
            if wordlist[i] not in self.stats and tmp not in visited:
                self.dfs_cas(wordlist[i],"",[wordlist[i][0].lower(),wordlist[i][0].upper()],0,voyel)
                visited.add(tmp)
        for i in range(len(wordlist)):
            tmp = self.trace(wordlist[i],voyel)
            if tmp not in self.v:
                self.v[tmp] = wordlist[i]
        for i in range(len(queries)):
            print(queries[i]," : ")
            if queries[i] in wordset:
                
                pass
            else:
                if queries[i] in self.stats:
                    queries[i] = self.stats[queries[i]]
                else:
                    key = ""
                    for x in queries[i]:
                        if x in voyel:
                            key+="_"
                        else:
                            key+=x.lower()
                    if key in self.v:
                        queries[i]=self.v[key]
                    else:
                        queries[i]="" 
            print(queries[i],"\n")
        return queries
    
    def trace(self,mot,voyel):
        res = ""
        for x in mot:
            if x in voyel:
                res+='_'
            else:
                res+=x.lower()
        return res
        
        
    def dfs_cas(self,mot,cur,choices,index,voyel):
        if index+1==len(mot):
            self.stats[cur+choices[0]]=mot
            self.stats[cur+choices[1]]=mot
            return
        for i in range(len(choices)):
            cur+=choices[i]
            choix = [mot[index+1].lower(),mot[index+1].upper()]
            self.dfs_cas(mot,cur,choix,index+1,voyel)
            cur = cur[:-1]
        return
    
x = Solution()
x.spellchecker(wordlist,queries)
```
